```python
# Define a functino to get data information
def getinfo(filename):
    config = configparser.ConfigParser()
    config.read(filename + '/config.ini')
    name = str(config['info']['name'])
    xvalue = config['info']['value_indices'].split(',')
    yvalue = config['info']['target_index']
    idindex = config['info']['id_indices']
    catindex = config['info']['categoric_indices'].split(',')
    header = config['info']['header_lines']
    separator = config['info']['separator']
    if separator == 'comma':
        sep = ','
    elif separator == '':
        sep = '\t'
    elif separator == ';':
        sep = ';'
    return {'name': name, 'xindex': xvalue, 'yindex': yvalue, 'idindex': idindex, 'catindex':catindex,
                  'header': header, 'separator': sep}
```

```python
# Load the dataset and prepared the data
from sklearn.preprocessing import LabelEncoder

def dataread(foldername, filename):
    info = getinfo(foldername)
    
    #read csv file
    try:
        Dataframe = pd.read_csv(foldername + '/' + filename, sep = info.get('separator'),skiprows = int(info.get('header')),
                    header = None, na_values = '?')
        
        if len(Dataframe.columns) == 1:
            Dataframe = pd.read_csv(foldername + '/' + filename, skiprows = int(info.get('header')), header = None,
                       na_values = '?', delim_whitespace = True)
        #len(Dataframe.colums)
    except:
        Dataframe = pd.read_csv(foldername + '/' + filename, skiprows = int(info.get('header')), header = None,
                       na_values = '?', delim_whitespace = True)
        
    Dataframe = Dataframe.dropna()
    
    #separate target and value variables
    
    l = getinfo(foldername)['xindex']
    m = getinfo(foldername)['yindex']
    xcolnum=[]
    for i in range(0,len(l)):
        xcolnum.append(int(l[i])-1)
    ycolnum = int(m)-1
    df_x = Dataframe.iloc[:, xcolnum]
    df_y = Dataframe.iloc[:, ycolnum]
    
    # data encoding/ transfer the categorical data
    
    encoder = LabelEncoder()
    data_cat_encoded = encoder.fit_transform(df_y)
    df_cat_encoded = pd.DataFrame(data_cat_encoded)
    data_y = pd.DataFrame(df_cat_encoded)
    
    data_dummy = pd.get_dummies(df_x)
    data_x = pd.DataFrame(data_dummy)
    
    return {'x': data_x, 'y': data_y}
```
